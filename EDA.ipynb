```python
# Importamos las librerías pertinentes.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Importamos los archivos csv procesados en el ETL.
penetracion_provincial = pd.read_csv('archivos/archivos_eda/penetracion_eda.csv')
penetracion_nacional = pd.read_csv('archivos/archivos_eda/penetracion_nacional_eda.csv')
tecnologia = pd.read_csv('archivos/archivos_eda/tecnologia_eda.csv')
telefonia_fija = pd.read_csv('archivos/archivos_eda/telefonia_fija_eda.csv')
telefonia_movil = pd.read_csv('archivos/archivos_enacom/telefonia_movil.csv')
television_paga = pd.read_csv('archivos/archivos_eda/television_paga_eda.csv')
velocidad = pd.read_csv('archivos/archivos_eda/velocidad_eda.csv')
```

```python
# Eliminamos la columna creada en la importación.
penetracion_provincial.drop('Unnamed: 0',axis=1,inplace=True)
penetracion_nacional.drop('Unnamed: 0',axis=1,inplace=True)
tecnologia.drop('Unnamed: 0',axis=1,inplace=True)
telefonia_fija.drop('Unnamed: 0',axis=1,inplace=True)
television_paga.drop('Unnamed: 0',axis=1,inplace=True)
velocidad.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
# Revertimos el orden de las columnas para darle una progresión de tiempo (pasado-futuro) a las visualizaciones.
penetracion_reversed = penetracion_provincial.iloc[::-1].reset_index()
penetracion_reversed.drop('index',axis=1,inplace=True)
penetracion_nacional_rev = penetracion_nacional.iloc[::-1].reset_index()
penetracion_nacional_rev.drop('index',axis=1,inplace=True)
tecnologia_rev = tecnologia.iloc[::-1].reset_index()
tecnologia_rev.drop('index',axis=1,inplace=True)
velocidad_rev = velocidad.iloc[::-1].reset_index()
velocidad_rev.drop('index',axis=1,inplace=True)
television_paga_rev = television_paga.iloc[::-1].reset_index()
television_paga_rev.drop('index',axis=1,inplace=True)
```

```python
# Visualizamos el acceso a internet a nivel nacional.
plt.figure(figsize=(8,6))
plt.plot(penetracion_nacional_rev['Accesos por cada 100 hogares'],label='Acceso por cada 100 hogares')
plt.plot(penetracion_nacional_rev['Accesos por cada 100 hab'],label='Acceso por cada 100 habitantes')
plt.xticks([])
plt.xlabel('Años: 2014-2022')
plt.legend()
plt.title('Acceso a Internet')
```

```python
# Visualizamos la evolución de las cuatro tecnologías más importantes de Internet a nivel nacional.
plt.figure(figsize=(10,6))
plt.plot(tecnologia_rev['Periodo'],tecnologia_rev['Fibra óptica'],label='Fibra optica')
plt.plot(tecnologia_rev['Cablemodem'],label='Cablemodem')
plt.plot(tecnologia_rev['Wireless'],label='Wireless')
plt.plot(tecnologia_rev['ADSL'],label='ADSL')
plt.xticks(rotation=45)
plt.legend()
plt.title('Evolución de Tecnologías en Argentina')
```

Tecnologías que incrementan sus consumidores: Fibra óptica y Cablemodem.

Tecnologías que disminuyen sus consumidores: ADSL.

Tecnologías que se mantienen igual: Wireless.

```python
# Hacemos un mapa de correlación entre las tecnologías
sns.heatmap(tecnologia_rev[['ADSL','Cablemodem','Fibra óptica','Wireless']].corr())
```

Vemos una alta correlación positiva entre Cablemodem y Fibra óptica, y a la vez, una alta correlación negativa de ambas tecnologías con ADSL.
Estás correlaciones se pueden explicar por la creciente necesidad de mayor ancho de banda en la velocidad de bajada de los consumidores.

```python
# Visualizamos el consumo de telefonía fija a nivel nacional.
plt.plot(telefonia_fija['Accesos telefonía fija (comercial)'],label='comercial')
plt.plot(telefonia_fija['Accesos telefonía fija (hogares)'],label='hogares')
plt.plot(telefonia_fija['Accesos telefonía fija (gobierno)'],label='gobierno')
plt.plot(telefonia_fija['Accesos telefonía fija (otros)'],label='otros')
plt.plot(telefonia_fija['Accesos telefonía fija (total)'],label='total')
plt.title('Consumo de Telefonía fija')
plt.xticks([])
plt.xlabel('Años: 2014 - 2022')
plt.legend()
```

Vemos un consumo decreciente del año 2014 al año 2022 en el total y en la rama más importante(hogares). No así en el sector gubernamental ni en el comercial.

```python
# Visualizamos la evolución del consumo de ADSL y Telefonía Fija a nivel nacional.
plt.figure(figsize=(8,6))
plt.plot(tecnologia_rev['ADSL'],label='ADSL')
plt.plot(telefonia_fija['Accesos telefonía fija (total)'],label='Telefonía Fija')
plt.legend()
plt.xticks([])
plt.title('Comparación entre el consumo de telefonía fija y el consumo de ADSL')
plt.xlabel('Del Año 2014 al 2022')
```

```python
# Visualizamos la evolución de la velocidad de bajada media (Mbps) a nivel nacional.
plt.figure(figsize=(4,2))
plt.plot(velocidad_rev['Mbps (Media de bajada)'])
plt.xticks([])
plt.xlabel('Años: 2014 a 2022')
plt.title('Velocidad media de bajada: Mbps')
```

```python
# Chequeamos niveles de correlación entre el consumo de la tecnología ADSL y el consumo de la telefonía fija.

np.corrcoef(tecnologia_rev['ADSL'].values,telefonia_fija['Accesos telefonía fija (total)'].values)
```

```python
# Chequeamos niveles de correlación entre el consumo de la tecnología ADSL y la velocidad media de bajada.

np.corrcoef(tecnologia_rev['ADSL'].values,velocidad_rev['Mbps (Media de bajada)'].values)
```

Hay una correlación fuerte entre el consumo de la tecnología ADSL y el consumo de telefonía fija y tiene sentido porque la tecnología ADSL usa el cableado de la telefonía fija para enviar su señal, por ende, muchas operadoras de telefonía fija arman paquetes de promoción con Internet + Telefonía Fija.

Vemos una correlación negativa muy fuerte entre el consumo de la tecnología ADSL y la velocidad media de bajada. Esto puede significar que, dado que la tecnología ADSL tiene un límite de velocidad de bajada situado en 24 Mbps, y las costumbres han ido llevándonos hacia el consumo de streaming(Netflix,Amazon Prime, Youtube,Spotify), el límite de la tecnología ADSL es muy bajo para las costumbres actuales.
(Netflix recomienda (3Mbps para 720p, 5MBps para 1080p y 15Mbps para 4k) https://help.netflix.com/es/node/306)

Si a eso le sumamos que el la vida actual pasa mucho por las redes sociales(Instagram, Facebook,Tik-tok,Twitter), y que los distintos dispositivos que consumen Internet se van sumando en el consumo simultáneo, se necesitan cada vez más velocidades más grandes de bajada.
